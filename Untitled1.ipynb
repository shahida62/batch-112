```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
x = np.arange(11)
y = x ** 2
print(x,y)
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y)
plt.xlabel('x axis = year')
plt.ylabel('y axis = price')
plt.title('Bangladeshi oil price 2000-2025')
plt.savefig('Bangladesh oil price.png')
plt.show
```

```python
plt.subplot(2,3,1)
plt.plot(y,x, 'g--')

plt.subplot(2,3,2)
plt.plot(y,x, 'b--')


plt.subplot(2,3,3)
plt.plot(y,x, 'r--')


plt.subplot(2,3,4)
plt.plot(y,x, 'b--')

plt.subplot(2,3,5)
plt.plot(y,x, 'b--')

plt.subplot(2,3,6)
plt.plot(y,x, 'b--')

```
