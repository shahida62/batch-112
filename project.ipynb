```python
import pandas as pd
```

```python
# Created a data
```

```python
data = {
    "Name": ["Asha", "Rafi", "Sinthiya", "Ammal", "Rabbi", "Jinia"],
    "Age": [18, 19, 17, 20, 15, 16],
    "Gender": ["Female", "Male", "Female", "Male", "Male", "Female"],
    "Subject": ["Bangla", "English", "Math", "Social Science", "Science", "Applied Math"],
    "Bangla_Score": [85, 55, 78, 80, 90, 95],
    "English_Score": [95, 80, 85, 70, 75, 99],
    "Math_Score": [100, 98, 96, 92, 89, 86],
    "SocialScience_Score": [87, 88, 93, 94, 98, 100]
}
```

```python
df = pd.DataFrame(data)
```

```python
# save to csv
```

```python
df.to_csv("Student.csv", index=False)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.title("Average Math Score by Gander")
```

```python
plt.xlabel("Average score")
```

```python
plt.ylabel("Average Score")
```

```python
plt.pie(df["Bangla_Score"], labels=df["Name"], autopct="%1.1f%%", startangle=90)
plt.title("Bangla Score Distribution")
plt.show()
```

```python
# Histogram(Score Distribution)
```

```python
plt.hist(df["SocialScience_Score"], bins=5, color="orange", edgecolor="black")
plt.title("Social Science Score Distribution")
plt.xlabel("Score")
plt.ylabel("Frequency")
plt.show()
```

```python
# Bar Plot
```

```python
sns.barplot(x="Name", y="Math_Score", data=df, palette="cool")
plt.title("Math Score By Student")
plt.show()
```

```python
# Box Plot(Detect Outliers)
```

```python
sns.boxplot(x="Gender", y="English_Score", data=df, palette="pastel")
plt.title("English Score by Gender")
plt.show()
```

```python
# Scatter plot (Relationship Between Two Variables)
```

```python

sns.scatterplot(x="Bangla_Score", y="English_Score", data=df, color="purple")
plt.title("Bangla vs English Score")
plt.xlabel("Bangla Score")
plt.ylabel("English Score")
plt.show()
```

```python
# Pair Plot(Multiple Relationship)
```

```python
sns.pairplot(df, hue="Gender")
plt.show()
```

```python
# Heatmap (Correlation)
```

```python
correlation = df[["Math_Score", "English_Score", "Age"]].corr()
sns.heatmap(correlation, annot=True, cmap="Blues")
plt.title("Correlation Between Score And Age")
plt.show()
```

```python
# Combine Pandas + Seaborn for Analysis
# Average Bangla Score By Gender
```

```python
avg = df.groupby("Gender")["Bangla_Score"].mean().reset_index()
sns.barplot(x="Gender", y="Bangla_Score", data=avg, palette="viridis")
plt.title("Average Bangla Score By Gender")
plt.show()
```
