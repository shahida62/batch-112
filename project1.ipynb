```python
import pandas as pd
```

```python
data = {
    "OrderID": [1001, 1002, 1003, 1004, 1005, 1006],
    "Date": ["2024-01-02",  "2024-01-03", "2024-01-04", "2024-01-05", "2024-01-06", "2024-01-07"],
    "Product": ["Notebook", "Pencil", "Mouse", "Keyboard", "Laptop", "Headphone"],
    "Category": ["Stationery", "Stationery", "Electronics", "Electronics", "Electronics", "Electronics"],
    "price": [50, 10, 600, 1500, 55000, 1000],
    "Quantity": [2, 5, 1, 1, 2, 1],
    "Total": [100, 50, 200, 600, 55000, 300],
    "Customer_Age": [23, 31, 27, 25, None, 45],
    "Payment_Method": ["Cash", "Card", "Online", "Cash", "Online", "Card"],
    
}
```

```python
df = pd.DataFrame(data)
```

```python
df.to_csv("Sales_data.csv", index=False)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Category_sales = df.groupby("Category")["Total"].sum().reset_index()

plt.figure(figsize=(8,5))
sns.barplot(data=Category_sales, x="Category", y="Total")
plt.title("Total Sales by Category")
plt.show()
```

```python
product_sales = df.groupby("Product")["Total"].sum().sort_values(ascending=False).head(5)
plt.figure(figsize=(8,5))
product_sales.plot(kind="bar")
plt.title("Top 5 Products by Sales")
plt.xlabel("Product")
plt.ylabel("Total Sales")
plt.show()
```

```python
payment_counts = df["Payment_Method"].value_counts()
plt.figure(figsize=(6,6))
payment_counts.plot(kind="pie", autopct="%1.1f%%")
plt.title("Payment Method Distribution")
plt.ylabel("Total Payment")
plt.show()
```

```python
plt.figure(figsize=(8,5)) 
plt.scatter(df["Customer_Age"], df["Total"])
plt.title("Customer Age vs Total Spending")
plt.xlabel("Age")
plt.ylabel("Total Purchase Amount")
plt.show()
```
