```python
import pandas as pd
```

```python
data = {
    "OrderID": [1001, 1002, 1003, 1004, 1005, None],
    "Date": ["2024-01-02", "2024-01-03", None, "2024-01-05", "2024-01-06", "2024-01-07"],
    "Product": ["Mouse", None, "Keyboard", "Loptop", "Headphone", "Printer"],
    "Category": ["Electronics", "Electronics", "Electronics", "Electronics", "Electronics", "Electronics"],
    "price": [500, 1000, 50000, 700, 15000, 3000],
    "Quantity": [2, 5, 3, 1, 6, 1],
    "Total": [100, 50, 600, 55000, 400, 700],
    "Customer_Age": [23, 31, 40, 50, 55, 60],
    "Payment_Method": ["Cash", "Card", "Online", "Cash", "Card", "Online"],


}
```

```python
df = pd.DataFrame(data)
```

```python
df.to_csv("Customer_Order_data.csv", index=False)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Visualization with matplotlib
```

```python
Category_sales = df.groupby("Category")["Total"].sum().reset_index()
plt.figure(figsize=(8,5))
sns.barplot(data=Category_sales, x="Category", y="Total")
plt.title("Total Sales by Category")
plt.show()
```

```python
product_sales = df.groupby("Product")["Total"].sum().sort_values(ascending=False).head(5)

plt.figure(figsize=(8,5))
product_sales.plot(kind="bar")   # FIXED
plt.title("Top 5 Products by Sales")
plt.xlabel("Product")
plt.ylabel("Total Sales")
plt.show()
```

```python
daily_sales = df.groupby("Date")["Total"].sum()
plt.figure(figsize=(10,5))
daily_sales.plot(kind="line", marker="o")
plt.title("Daly Sales Trend")
plt.xlabel("Date")
plt.ylabel("Total Sales")
plt.grid(True)
plt.show()
```

```python
payment_counts = df["Payment_Method"].value_counts()

plt.figure(figsize=(6,6))
payment_counts.plot(kind="pie", autopct="%1.1f%%")
plt.title("Payment Method Distribution")
plt.ylabel("")   # optional  hides the y-label
plt.show()
```
